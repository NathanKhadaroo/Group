### Loading packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
import datetime
import math

from sklearn.preprocessing import MinMaxScaler
```

### Importing the files

```python
db_store = pd.read_csv("DA1920_store.csv")
db_train = pd.read_csv("DA1920_train.csv")
db_test = pd.read_csv("DA1920_test.csv")
```

### Merging the datasets

```python
dbj1 = db_train.join(db_store.set_index("Store"), on = "Store")
dbj1['Index'] = dbj1.index
dbj1 = dbj1.set_index('Index')
dbj1.head()
```

```python
len(dbj1.index)
```

### Deleting the 'unnamed' columns

```python
dbj1 = dbj1.drop(columns = ['Unnamed: 10', 'Unnamed: 11'])
dbj1.head()
```

### Counting missing data points in every column

```python
dbj1.isnull().sum()
```

### Creating new variables for competition since and promo 2 since

```python
dbj1['Year'] = pd.DatetimeIndex(dbj1['Date']).year
dbj1['month'] = pd.DatetimeIndex(dbj1['Date']).month
dbj1['week'] = pd.DatetimeIndex(dbj1['Date']).week

dbj1["Had_competitor_since_months"] = (((dbj1["Year"]- dbj1["CompetitionOpenSinceYear"])*12)\
                                        + (dbj1["month"]- dbj1["CompetitionOpenSinceMonth"]))

dbj1["Had_promo2_since_weeks"] = (((dbj1["Year"]- dbj1["Promo2SinceYear"])*52.143)\
                                        + (dbj1["week"]- dbj1["Promo2SinceWeek"]))
dbj1.head()
```

This creates some negative values as values are given even for future competition openings and promo begginings. We can NaN these values and still have our new variables. 

```python
dbj1.loc[(dbj1['Had_promo2_since_weeks'] < 0)] = math.nan
dbj1.loc[(dbj1['Had_competitor_since_months'] < 0)] = math.nan
```

### Creating dummy variables

#### For time variables

- Creating dummies for the day of the week, default is Monday

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['DayOfWeek'], drop_first = True)], axis = 1)
dbj1 = dbj1.drop(columns = ['DayOfWeek'])
dbj1.rename(columns = {2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday',7:'Sunday'}, inplace=True)
dbj1.head()
```

- Creating dummies for the month, default is January

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['month'],drop_first = True)], axis = 1)
dbj1 = dbj1.drop(columns = ['month'])
dbj1.rename(columns = {2:'Feb.', 3: 'Mar.', 4: 'Apr.', 5: 'May', 6: 'June', 7: 'July', 8: 'Aug.', 9: 'Sept.', 10:'Oct.', 11: 'Nov.', 12: 'Dec'}, inplace=True)
dbj1.columns
```

- Creating dummies for the year, default is 2013

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['Year'],drop_first = True)], axis = 1)
dbj1 = dbj1.drop(columns = ['Year'])
dbj1.columns
```

#### For other variables

- Creating dummies for the Assortment available, default is a

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['Assortment'],drop_first = True)], axis = 1)
dbj1 = dbj1.drop(columns = ['Assortment'])
dbj1.rename(columns = {'b': 'Assortment_b', 'c': 'Assortment_c'}, inplace=True)
dbj1.columns
```

- Creating dummies for the Promo Interval, default is no promo

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['PromoInterval'])], axis = 1)
dbj1 = dbj1.drop(columns = ['PromoInterval'])
dbj1.columns
```

- Creating dummies for State Holidays, the default is no holiday

```python
pd.unique(dbj1['StateHoliday'])
```

Some variables are labelled 0 and other '0', we need to make them the same.

```python
dbj1.loc[(dbj1['StateHoliday'] == 0)] = '0'
```

```python
dbj1 = pd.concat([dbj1, pd.get_dummies(dbj1['StateHoliday'])], axis = 1)
dbj1 = dbj1.drop(columns = ['StateHoliday', '0'])
dbj1.rename(columns = {'a': 'Public_Holiday','b': 'Easter holiday', 'c': 'Christmas'}, inplace=True)
dbj1.columns
```

### Standardising the data

```python
variables = ['Sales','Customers','CompetitionDistance']
scaler = MinMaxScaler() 
db_scaled = scaler.fit_transform(dbj1[variables])
x = pd.DataFrame(db_scaled,
                 columns = variables
                )
x.rename(columns = {'Sales': 'Sales_norm','Customers':'Customers_norm','CompetitionDistance':'CompetitionDistance_norm'}, inplace=True)
x['Index'] = x.index
x = x.set_index('Index')
dbj1 = dbj1.join(x, on = 'Index')
dbj1.head()
```

### Le Heatmap de Laura

```python
k = 30
#create a list called cols. that looks for the 10 largest correlations to SalePrice in the dataset . 
cols = dbj1.corr().nlargest(k, 'Sales')['Sales'].index
#create a correlation matrix using the list of columns created above. 
k_corr_matrix = dbj1[cols].corr()
#plot the figure with given dimensions. 
plot.figure(figsize=(48, 32))
#display the correlation matrix using seaborn's heatmap function. plot strength from red to blue.
sns.heatmap(k_corr_matrix, annot=True, cmap=plot.cm.RdBu_r)
```
